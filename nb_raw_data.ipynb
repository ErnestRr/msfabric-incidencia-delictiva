```python
from pyspark.sql.functions import col
import re

# Rutas de los archivos
#la ruta debe apuntar directamente a tu archivo municipal.csv
csv_file_path = "Files/raw/delitos_municipales/municipal.csv"
output_table_name = "delitos_municipales_silver"

# Leer el archivo CSV con la codificación correcta
print("Leyendo el archivo CSV...")
try:
    # Lee el archivo CSV con la codificación 'latin1'
    df = spark.read.option("header", "true").option("inferSchema", "true").option("encoding", "latin1").csv(csv_file_path)
    print(" Archivo leído exitosamente.")
except Exception as e:
    print(f" Error al leer el archivo. Verifica la ruta y el nombre del archivo. Detalles: {e}")
    exit()

# Limpiar los nombres de las columnas
print("\nLimpiando nombres de columnas...")
old_columns = df.columns
new_columns = [re.sub(r'[^a-zA-Z0-9_]', '_', column).strip('_').lower() for column in old_columns]
df_cleaned = df.toDF(*new_columns)
print(" Nombres de columnas limpios.")

# Guardar el DataFrame limpio en una tabla Delta en la capa silver
print("\nGuardando los datos limpios en la capa silver...")
try:
    # Usa saveAsTable para crear una tabla SQL en tu Lakehouse
    df_cleaned.write.format("delta").mode("overwrite").saveAsTable(output_table_name)
    print(f" ¡Datos guardados exitosamente en la tabla Delta '{output_table_name}'! ")
except Exception as e:
    print(f" Error al guardar la tabla. Detalles: {e}")
```

```python
# Crear un DataFrame con las columnas renombradas para la tabla de hechos
hechos_delitos = spark.sql(
    """
    SELECT
        T1.id_modalidad,
        T2.id_bien_juridico,
        T3.id_tipo_delito,
        T4.id_subtipo_delito,
        T5.clave_ent,
        T5.cve__municipio,
        T6.Valor AS incidencias,
        CONCAT_WS('/', '01',
            CASE T6.Atributo
                WHEN 'Enero' THEN '01'
                WHEN 'Febrero' THEN '02'
                WHEN 'Marzo' THEN '03'
                WHEN 'Abril' THEN '04'
                WHEN 'Mayo' THEN '05'
                WHEN 'Junio' THEN '06'
                WHEN 'Julio' THEN '07'
                WHEN 'Agosto' THEN '08'
                WHEN 'Septiembre' THEN '09'
                WHEN 'Octubre' THEN '10'
                WHEN 'Noviembre' THEN '11'
                WHEN 'Diciembre' THEN '12'
                ELSE T6.Atributo
            END,
        T6.a_o) AS fecha
    FROM delitos_municipales_bronze AS T6
    LEFT JOIN lh_silver.dim_modalidad AS T1 ON T6.Modalidad = T1.modalidad
    LEFT JOIN lh_silver.dim_bien_juridico AS T2 ON T6.bien_jur_dico_afectado = T2.bien_juridico_afectado
    LEFT JOIN lh_silver.dim_tipo_delito AS T3 ON T6.tipo_de_delito = T3.tipo_de_delito
    LEFT JOIN lh_silver.dim_subtipo_delito AS T4 ON T6.subtipo_de_delito = T4.subtipo_de_delito
    LEFT JOIN lh_silver.dim_entidad AS T5 ON T6.clave_ent = T5.clave_ent AND T6.cve__municipio = T5.cve__municipio
    """
)

# Guardamos la tabla de hechos en la capa Silver
hechos_delitos.write.format("delta").mode("overwrite").saveAsTable("lh_silver.hechos_incidencias")
```

```python
from pyspark.sql.functions import col, concat_ws, when, lit

# Cargar el DataFrame de la tabla de origen
df_bronce = spark.table("delitos_municipales_bronze")

# Crear el DataFrame para la dimensión de calendario con mes, año y una columna de tiempo
dim_calendario = df_bronce.select(
    col("a_o").alias("annio"),
    col("Atributo").alias("mes"),
).distinct()

# Crear la columna 'time' combinando el mes y el año
dim_calendario = dim_calendario.withColumn(
    "time",
    concat_ws('-',
        when(col("mes") == "Enero", lit("01"))
        .when(col("mes") == "Febrero", lit("02"))
        .when(col("mes") == "Marzo", lit("03"))
        .when(col("mes") == "Abril", lit("04"))
        .when(col("mes") == "Mayo", lit("05"))
        .when(col("mes") == "Junio", lit("06"))
        .when(col("mes") == "Julio", lit("07"))
        .when(col("mes") == "Agosto", lit("08"))
        .when(col("mes") == "Septiembre", lit("09"))
        .when(col("mes") == "Octubre", lit("10"))
        .when(col("mes") == "Noviembre", lit("11"))
        .when(col("mes") == "Diciembre", lit("12"))
        .otherwise(col("mes")),
        col("annio")
    )
)

# Guardar la tabla de dimensión en la capa Silver
dim_calendario.write.format("delta").mode("overwrite").saveAsTable("lh_silver.dim_calendario")

# Opcional: mostrar las primeras filas de la tabla creada para verificar
dim_calendario.show(10, truncate=False)
```
